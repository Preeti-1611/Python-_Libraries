```python
#iterating with  nditer()
arr = np.array([[1,2],[3,4]])
```

```python
for x in np.nditer(arr):
    print(x,end = " ")
```

```python
# ndenumerate() -> both index +value
for ind, x in np.ndenumerate(arr):
    print(ind, x)
```

```python
arr = np.array([[1,2],[3,4]])
# np.nditer is an efficient multi-dimensional iterator in NumPy.
# It allows you to iterate through every element of an array, 
# regardless of how many dimensions it has.
for x in np.nditer(arr):
    print(x)
```

```python
arr = np.array([1,2,3])

v = arr.view()
c = arr.copy()

v[0] = 99
c[1] = 100

print(arr)  # only view affected
```

```python
# np.ndenumerate allows iterating over an array while returning both the element’s index and value.
# It is useful when you need both the position and the item, especially in multi-dimensional arrays.”
import numpy as np

arr = np.array([[10, 20],
                [30, 40]])

for index, value in np.ndenumerate(arr):
    print(index, value)
```
