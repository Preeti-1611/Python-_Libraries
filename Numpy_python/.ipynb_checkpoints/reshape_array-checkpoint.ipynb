```python
import numpy as np
#array reshape
arr = np.array([1,2,3,4,5,6])
reshaped = arr.reshape(2,3)
print(reshaped)
```

```python
reshape_2 = reshaped.reshape(3,2)
reshape_2
```

```python
#revel -> convert 1d array
ravel = reshape_2.ravel()
print(ravel)
ravel[0]=100
print(ravel)
print(reshape_2)
```

```python
#flatten -> to 1d array
# returns a copy of the array
flat = reshape_2.flatten()
flat
```

```python
flat[0]=1
```
