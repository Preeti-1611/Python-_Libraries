```python
import numpy as np

# String array
arr_str = np.array(["10", "20", "30"])
print("Before:", arr_str, arr_str.dtype)

# Convert string to int
arr_int = arr_str.astype(int)
print("After:", arr_int, arr_int.dtype)
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50], dtype=np.int64)

print("Array:", arr)
print("Size:", arr.size)         # Number of elements
print("Itemsize:", arr.itemsize) # Bytes per element
print("Total bytes (nbytes):", arr.nbytes)  # Total memory used
```
