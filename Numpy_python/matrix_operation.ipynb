```python
import numpy as np

A = np.array([[1,2,3],
              [4,5,6]])       # shape (2,3)

print("A shape:", A.shape)
print("A:\n", A)

At = A.T                    # transpose
print("\nA.T shape:", At.shape)
print("A.T:\n", At)
```

```python
#swapaxes -> swap 2 specific axes
arr = np.array([[[1,2],[3,4]]])
arr.shape
```

```python
swap = np.swapaxes(arr,0,1)
swap.shape
```

```python
A = np.array([[1,2,3],
              [4,5,6]])   # shape (2,3)
print(np.swapaxes(A, 0, 1).shape)
```

```python
#concatination
a = np.array([1,2])
b = np.array([3,4])
```

```python
#vertical stack
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6],[7,8]])
```

```python
print(np.vstack((arr1,arr2)))
```

```python
#h=-stack
print(np.hstack((arr1,arr2)))
```

```python
print(np.stack((arr1,arr2),axis =0))
```

```python
print(np.stack((arr1,arr2),axis =1))
```
